### January 2023

```python
jan_23 = pd.read_csv(r'data/JC-202301-citibike-tripdata.csv', na_filter=False, parse_dates=['started_at', 'ended_at'])
month = cycle(['January 2023'])
jan_23['Month'] = [next(month) for m in range(len(jan_23))]

jan_23.sample(5, random_state = 123)
```

```python
print(jan_23.isnull().sum())

print('\n' +  '*' * 50, '\nInfo')
print('*' * 50)
jan_23.info()
```

```python
jan_23.dropna(inplace = True)
jan_23.drop(columns=['ride_id'], axis=1, inplace=True)
jan_23.reset_index(inplace = True, drop = True)

trip_duration_jan = []
for i in range(len(jan_23)):
    trip_duration_jan.append((int(jan_23.ended_at[i].hour) * 3600 + int(jan_23.ended_at[i].minute) * 
                              60 + int(jan_23.ended_at[i].second)) -
                             (int(jan_23.started_at[i].hour) * 3600 + int(jan_23.started_at[i].minute) *
                              60 + int(jan_23.started_at[i].second)))

jan_23['tripduration'] = trip_duration_jan
jan_23['day_of_week'] = jan_23['started_at'].dt.dayofweek
jan_23['number_of_records'] = 1

jan_23['day_of_week'] = jan_23['day_of_week'].map({ 0: 'Monday', 1: 'Tuesday', 2: 'Wednesday',
                                                   3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'})

jan_23.sample(5, random_state=123)
```

```python
start_station = []
end_station   = []

for i in range(jan_23.shape[0]):
    end_station.append(jan_23.end_station_id[i][2:])
    start_station.append(jan_23.start_station_id[i][2:])

jan_23['numeric_end_station_id']   = end_station
jan_23['numeric_start_station_id'] = start_station
jan_23['numeric_end_station_id']   = jan_23.numeric_end_station_id.astype('int64')
jan_23['numeric_start_station_id'] = jan_23.numeric_start_station_id.astype('int64')

jan_23.start_lat = jan_23.start_lat.astype(float)
jan_23.start_lng = jan_23.start_lng.astype(float)
jan_23.end_lat   = jan_23.end_lat.astype(float)
jan_23.end_lng   = jan_23.end_lng.astype(float)

print('\n' +  '*' * 50, '\nInfo')
print('*' * 50)
jan_23.info()
```
