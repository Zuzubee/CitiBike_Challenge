### July 2021

```python
jul_21 = pd.read_csv('data/2021/JC-202107-citibike-tripdata.csv')
month = cycle(['July 2021'])
jul_21['Month'] = [next(month) for m in range(len(jul_21))]

jul_21['started_at'] = jul_21['started_at'].apply(pd.to_datetime)
jul_21['ended_at'] = jul_21['ended_at'].apply(pd.to_datetime)

jul_21['started_at'] = jul_21['started_at'].dt.strftime('%m-%d-%Y %H:%M:%S')
jul_21['ended_at'] = jul_21['ended_at'].dt.strftime('%m-%d-%Y %H:%M:%S')

jul_21['started_at'] = jul_21['started_at'].apply(pd.to_datetime)
jul_21['ended_at'] = jul_21['ended_at'].apply(pd.to_datetime)

trip_duration_jul = []
for i in range(len(jul_21)):
    trip_duration_jul.append((int(jul_21.ended_at[i].hour) * 3600 + int(jul_21.ended_at[i].minute) * 
                              60 + int(jul_21.ended_at[i].second)) -
                             (int(jul_21.started_at[i].hour) * 3600 + int(jul_21.started_at[i].minute) *
                              60 + int(jul_21.started_at[i].second)))

jul_21['tripduration'] = trip_duration_jul

jul_21.head()
```

```python
start_station = []
end_station   = []

for i in range(jul_21.shape[0]):
    end_station.append(jul_21.start_station_id[i][2:])
    start_station.append(jul_21.start_station_id[i][2:])

jul_21.end_station_id   = end_station
jul_21.start_station_id = start_station

jul_21.end_station_id   = jul_21.end_station_id.astype(int)
jul_21.start_station_id   = jul_21.start_station_id.astype(int)

jul_21.drop(columns=['ride_id'], axis=1, inplace=True)

jul_21.info()
```

```python
jul_21.sample(5)
```
